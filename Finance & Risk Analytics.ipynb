```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() 
import warnings
warnings.filterwarnings("ignore")
pd.set_option('display.max_columns', None)
```

```python
index = pd.read_csv('S&P500.csv',usecols=['Date','Close','Volume'])
index.columns = ['Date','S&P Close', 'Volume']
```

```python
AAPL =  pd.read_csv('AAPL.csv',usecols=['Date','Close','Volume'])
AAPL.columns=['Date','tech_AAPL_close','tech_AAPL_volume']
AMZN = pd.read_csv('AMZN.csv',usecols=['Date','Close','Volume'])
AMZN.columns=['Date','tech_AMZN_close','tech_AMZN_volume']
FB =   pd.read_csv('FB.csv',usecols=['Date','Close','Volume'])
FB.columns=['Date','tech_FB_close','tech_FB_volume']
GOOG = pd.read_csv('GOOG.csv',usecols=['Date','Close','Volume'])
GOOG.columns=['Date','tech_GOOG_close','tech_GOOG_volume']
IBM  = pd.read_csv('IBM.csv',usecols=['Date','Close','Volume'])
IBM.columns=['Date','tech_IBM_close','tech_IBM_volume']
MFST = pd.read_csv('MSFT.csv',usecols=['Date','Close','Volume'])
MFST.columns=['Date','tech_MFST_close','tech_MFST_volume']
```

```python
BHC =  pd.read_csv('BHC.csv',usecols=['Date','Close','Volume'])
BHC.columns=['Date','pharma_BHC_close','pharma_BHC_volume']
JNJ = pd.read_csv('JNJ.csv',usecols=['Date','Close','Volume'])
JNJ.columns=['Date','pharma_JNJ_close','pharma_JNJ_volume']
MRK =   pd.read_csv('MRK.csv',usecols=['Date','Close','Volume'])
MRK.columns=['Date','pharma_MRK_close','pharma_MRK_volume']
PFE = pd.read_csv('PFE.csv',usecols=['Date','Close','Volume'])
PFE.columns=['Date','pharma_PFE_close','pharma_PFE_volume']
RHHBY  = pd.read_csv('RHHBY.csv',usecols=['Date','Close','Volume'])
RHHBY.columns=['Date','pharma_RHHBY_close','pharma_RHHBY_volume']
UNH = pd.read_csv('UNH.csv',usecols=['Date','Close','Volume'])
UNH.columns=['Date','pharma_UNH_close','pharma_UNH_volume']
```

```python
BCS =  pd.read_csv('BCS.csv',usecols=['Date','Close','Volume'])
BCS.columns=['Date','finance_BCS_close','finance_BCS_volume']
CS = pd.read_csv('CS.csv',usecols=['Date','Close','Volume'])
CS.columns = ['Date','finance_CS_close','finance_CS_volume']
DB =   pd.read_csv('DB.csv',usecols=['Date','Close','Volume'])
DB.columns = ['Date','finance_DB_close','finance_DB_volume']
GS = pd.read_csv('GS.csv',usecols=['Date','Close','Volume'])
GS.columns = ['Date','finance_GS_close','finance_GS_volume']
MS  = pd.read_csv('MS.csv',usecols=['Date','Close','Volume'])
MS.columns = ['Date','finance_MS_close','finance_MS_volume']
WFC = pd.read_csv('WFC.csv',usecols=['Date','Close','Volume'])
WFC.columns = ['Date','finance_WFC_close','finance_WFC_volume']
```

```python
AAL =  pd.read_csv('AAL.csv',usecols=['Date','Close','Volume'])
AAL.columns = ['Date','aviation_AAL_close','aviation_AAL_volume']
ALGT = pd.read_csv('ALGT.csv',usecols=['Date','Close','Volume'])
ALGT.columns = ['Date','aviation_ALGT_close','aviation_ALGT_volume']
ALK =   pd.read_csv('ALK.csv',usecols=['Date','Close','Volume'])
ALK.columns = ['Date','aviation_ALK_close','aviation_ALK_volume']
DAL = pd.read_csv('DAL.csv',usecols=['Date','Close','Volume'])
DAL.columns = ['Date','aviation_DAL_close','aviation_DAL_volume']
HA  = pd.read_csv('HA.csv',usecols=['Date','Close','Volume'])
HA.columns = ['Date','aviation_HA_close','aviation_HA_volume']
LUV = pd.read_csv('LUV.csv',usecols=['Date','Close','Volume'])
LUV.columns =['Date','aviation_LUV_close','aviation_LUV_volume']
```

```python
tech = pd.merge(AAPL,AMZN,how="outer",on="Date")
t = [FB,GOOG,IBM,MFST]
for i in t:
    tech = tech.merge(i,how="outer",on="Date")
```

```python
tech.Date=pd.to_datetime(tech.Date)
```

```python
tech.head()
```

```python
pharma = pd.merge(BHC,JNJ,how="outer",on="Date")
p = [MRK,PFE,RHHBY,UNH]
for i in p:
    pharma = pharma.merge(i,how="outer",on="Date")
```

```python
pharma.Date = pd.to_datetime(pharma.Date)
```

```python
pharma.head()
```

```python
finance = pd.merge(BCS,CS,how="outer",on="Date")
f = [DB,GS,MS,WFC]
for i in f:
    finance = finance.merge(i,how="outer",on="Date")
    
finance.Date = pd.to_datetime(finance.Date)
```

```python
finance.head()
```

```python
aviation = pd.merge(AAL,ALGT,how="outer",on="Date")
a=[ALK,DAL,HA,LUV]
for i in a:
    aviation = aviation.merge(i,how="outer",on="Date")
aviation.Date = pd.to_datetime(aviation.Date)
aviation.head()
```

```python
pharma.drop('Date',axis=1,inplace=True)
finance.drop('Date',axis=1,inplace=True)
aviation.drop('Date',axis=1,inplace=True)
index.drop('Date',axis=1,inplace=True)
```

```python
df = pd.concat([tech,pharma,finance,aviation,index],axis=1)
```
